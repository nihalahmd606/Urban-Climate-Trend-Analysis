```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12,6)
plt.rcParams['font.family'] = 'DejaVu Sans'
TITLE_FONT = {
    'fontsize': 20,
    'fontweight': 'bold'}

LABEL_FONT = {
    'fontsize': 13}
```

```python
plt.figure(figsize=(14,8))
corr = ab.corr(numeric_only=True)
sns.heatmap(
    corr,
    annot=True,
    cmap='Spectral',
    linewidths=0.5,
    fmt='.2f')

plt.title(
    "CLIMATE CORRELATION MATRIX\nUNDERSTANDING RELATIONSHIPS BETWEEN WEATHER VARIABLES",
    **TITLE_FONT)

plt.tight_layout()
plt.show()
```

```python
top_hot = (
    ab.groupby('city')['avg_temp']
    .mean()
    .sort_values(ascending=False)
    .head(10))

plt.figure(figsize=(14,7))

sns.barplot(
    x=top_hot.values,
    y=top_hot.index,
    palette='rocket')

plt.title(
    "TOP 10 HOTTEST CITIES WORLDWIDE\nURBAN AREAS FACING EXTREME HEAT",
    **TITLE_FONT
)

plt.xlabel("Average Temperature (°C)", **LABEL_FONT)
plt.ylabel("City", **LABEL_FONT)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(
    data=ab,
    x='humidity',
    y='heat_index',
    hue='heat_index',
    palette='coolwarm',
    s=100,
    alpha=0.7
)

plt.title(
    "HUMIDITY vs HEAT INDEX\nHOW MOISTURE INTENSIFIES HUMAN HEAT PERCEPTION",
    **TITLE_FONT
)

plt.xlabel("Humidity (%)", **LABEL_FONT)
plt.ylabel("Heat Index", **LABEL_FONT)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.histplot(
    ab['avg_temp'],
    bins=30,
    kde=True,
    color='purple'
)

plt.title(
    "DISTRIBUTION OF URBAN TEMPERATURE\nANALYZING CLIMATE SPREAD ACROSS CITIES",
    **TITLE_FONT
)

plt.xlabel("Average Temperature (°C)", **LABEL_FONT)
plt.ylabel("Frequency", **LABEL_FONT)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.boxplot(data=ab,
    x=ab.columns[0],
    y=ab.columns[1],
    palette='Set2')

plt.xticks(rotation=45)
plt.title(
    "TEMPERATURE VARIABILITY ACROSS WEATHER CONDITIONS\nIDENTIFYING EXTREME CLIMATE PATTERNS",
    fontsize=20,
    fontweight='bold')

plt.xlabel(
    "Weather Condition",
    fontsize=13)

plt.ylabel(
    "Average Temperature (°C)",
    fontsize=13)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,7))

plt.scatter(
    ab['humidity'],
    ab['avg_temp'],
    s=ab['humidity']*2,
    alpha=0.6,
    c=ab['heat_index'],
    cmap='turbo')

plt.colorbar(label='Heat Index')

plt.title(
    "AIR POLLUTION vs URBAN TEMPERATURE\nEXPLORING ENVIRONMENTAL STRESS IN CITIES",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel(
    "Air Quality Index",
    fontsize=13
)

plt.ylabel(
    "Average Temperature (°C)",
    fontsize=13
)

plt.tight_layout()
plt.show()
```

```python
from warnings
import warnings('ignore')
plt.figure(figsize=(14,7))

sns.violinplot(
    data=ab,
    x='wind_degree', 
    y='precip_mm',    
    palette='viridis'
)

plt.xticks(rotation=45)

plt.title(
    "Rainfall Distribution Across Weather Conditions\nUnderstanding Precipitation Intensity Patterns",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel(
    "Weather Condition",
    fontsize=13
)

plt.ylabel(
    "Precipitation (mm)",
    fontsize=13
)

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(3, 2, figsize=(20, 18))

fig.suptitle(
    "Urban Climate Trends Analysis Dashboard",
    fontsize=28,
    fontweight='bold')

top_hot = (
    ab.groupby('city')['avg_temp']
    .mean()
    .sort_values(ascending=False)
    .head(10))

sns.barplot(
    x=top_hot.values,
    y=top_hot.index,
    palette='rocket',
    ax=axes[0,0]) 

axes[0,0].set_title(
    "Top 10 Hottest Cities",
    fontsize=16,
    fontweight='bold')

axes[0,0].set_xlabel("Average Temperature (°C)")
axes[0,0].set_ylabel("City")

sns.scatterplot(
    data=ab,
    x='humidity',
    y='heat_index',
    hue='heat_index',
    palette='coolwarm',
    s=80,
    alpha=0.7,
    ax=axes[0,1])

axes[0,1].set_title(
    "Humidity vs Heat Index",
    fontsize=16,
    fontweight='bold')

sns.histplot(
    ab['avg_temp'],
    kde=True,
    bins=30,
    color='purple',
    ax=axes[1,0])

axes[1,0].set_title(
    "Temperature Distribution",
    fontsize=16,
    fontweight='bold')

ab['avg_temp'].plot(kind='box', ax=axes[1,1])

axes[1,1].set_title(
    "Temperature Distribution Box Plot", 
    fontsize=16,
    fontweight='bold')

axes[1,1].set_ylabel("Average Temperature (°C)")

corr = ab.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap='Spectral',
    fmt='.2f',
    ax=axes[2,0])

axes[2,0].set_title(
    "Climate Correlation Matrix",
    fontsize=16,
    fontweight='bold')

bubble = axes[2,1].scatter(
    ab['air_quality_index'],
    ab['avg_temp'],
    s=ab['humidity'] * 2,
    c=ab['heat_index'],
    cmap='turbo',
    alpha=0.6)

axes[2,1].set_title(
    "Air Quality vs Temperature",
    fontsize=16,
    fontweight='bold')

axes[2,1].set_xlabel("Air Quality Index")
axes[2,1].set_ylabel("Average Temperature (°C)")

fig.colorbar(
    bubble,
    ax=axes[2,1],
    label='Heat Index')

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
